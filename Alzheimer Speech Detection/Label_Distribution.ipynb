```python
# =============================
# Step 0 — Imports
# =============================
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# =============================
# Step 1 — Load dataset
# =============================
CSV_PATH = "D:/University/AI/Project/Alzheimer Speech Detection/data/addetector_dataset.csv"
df = pd.read_csv(CSV_PATH)
df = df.dropna()  # remove missing rows

# Map numeric labels to readable text
label_map = {0: "Healthy", 1: "Alzheimer"}
df['label_text'] = df['label'].map(label_map)
```

```python
# =============================
# Step 2 — Plot Label Distribution
# =============================
RESULTS_DIR = "D:/University/AI/Project/Alzheimer Speech Detection/results"
os.makedirs(RESULTS_DIR, exist_ok=True)

label_counts = df['label_text'].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=label_counts.index, y=label_counts.values, palette='Set2')
plt.title("Label Distribution")
plt.xlabel("Label")
plt.ylabel("Number of Samples")
plt.tight_layout()
plt.savefig(os.path.join(RESULTS_DIR, "label_distribution.pdf"))
plt.show()
```
