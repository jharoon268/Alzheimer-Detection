```python
import os
import librosa
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
AUDIO_DIR = "D:/University/AI/Project/Audio_for_Project/train_wavFiles/19-227"
durations = []
RESULTS_DIR = "D:/University/AI/Project/Alzheimer Speech Detection/results"

for file in os.listdir(AUDIO_DIR):
    if file.endswith(".wav"):
        path = os.path.join(AUDIO_DIR, file)
        y, sr = librosa.load(path, sr=None)
        durations.append(len(y) / sr)

# Plot histogram
plt.figure(figsize=(6,4))
sns.histplot(durations, bins=30, kde=True, color='skyblue')
plt.title("Audio Duration Histogram")
plt.xlabel("Duration (seconds)")
plt.ylabel("Number of Samples")
plt.savefig(os.path.join(RESULTS_DIR, "duration_histogram.pdf"))
plt.show()
```
