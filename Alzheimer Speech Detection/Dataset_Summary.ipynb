```python
import pandas as pd
```

```python
# Load the dataset CSV
CSV_PATH = "D:/University/AI/Project/Alzheimer Speech Detection/data/addetector_dataset.csv"
df = pd.read_csv(CSV_PATH)

# Drop missing rows
df = df.dropna()

# Total samples and features
num_samples = df.shape[0]
num_features = df.shape[1] - 1  # assuming last column is 'label'

# Map numeric labels to text
label_map = {0: "Healthy", 1: "Alzheimer"}
df['label_text'] = df['label'].map(label_map)

# Label counts
label_counts = df['label_text'].value_counts()

# Print summary
print("===== DATASET SUMMARY =====")
print(f"Total samples: {num_samples}")
print(f"Number of features per sample: {num_features}")
print("\nNumber of samples per label:")
print(label_counts)

# Optional: basic statistics of numeric columns
print("\nBasic statistics of numeric columns:")
print(df.describe())
```
